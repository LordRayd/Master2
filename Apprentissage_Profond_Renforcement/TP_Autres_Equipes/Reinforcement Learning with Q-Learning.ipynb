#Creation et exploration de l'environnement
Nous allons créer l'environnement grâce à OpenAI Gym à l'aide de la commande suivante :

```
env = gym.make("FrozenLake-v0")
```

Nous pouvons ensuite l'initialiser (reset). afficher la carte avec env.render() et effectuer des actions avec env.step()
step() prends un argument l'action à éffectuer (0,1,2,3 pour les différentes directions dans notre contexte)

env.action_space.sample() permet de choisir une action aléatoirement


```
env.reset()
env.render()
env.step(2) 
env.render()
env.step(env.action_space.sample()) 
env.render()
env.close()

new_state, reward, done, info = env.step(2)
env.render()
print(done)
```

à noter que la méthode step() renvois 4 valeurs :
- Le nouvel état atteint
- Récompense obtenu à cet état
- Un booléan contenant l'information de si l'épisode est terminé ou non (état final provoqué par la mort ou atteinte de l'objectif)
- un dictionnaire contenant des informations (probabilités...)

Il s'agit donc maintenant d'implémenter l'algorythme (la boucle principal) qui va pour rappel :
- Choisir une action parmis les différents
- Effectuer l'action et calculer la récompense
- Mettre à jour la Q-Value relatif à l'action choisie pour un état donné dans la Q-table

New Q-Value = <br>
Old Q-Value + Learning Rate \* [Reward + discount_rate\*(higher Q-value between possible actions from the newt state s')-Old Q-value] <br><br>
$NewQ(s,a)= Q(s,a) + \alpha [R(s,a)+\gamma maxQ'(s',a')-Q(s,a)]$
